```python
import pandas as pd
import matplotlib.pyplot as plt

# Data inlezen
df = pd.read_csv('meting_9.csv')

# Grafiek grootte aanpassen
plt.figure(figsize=(12, 6))

# Data plotten
plt.plot(df['Tijd (s)'], df['Spanning (V)'], label='Voltage Signaal')
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.title('Oscilloscoop Data - Béta verval meting 1')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
